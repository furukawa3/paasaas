# gitlabの設定

## 初期化

### パスワードの初期設定
初期パスワードが求められるので入力します。



### 管理者でサインイン
管理者で入力したパスワードを使ってサインインします。



* Username or email: root
* Password: 初期設定で入力したパスワード

### 正常性確認
サインインが成功することを確認します。




### サインアウト

管理者ユーザでサインアウトします。



## ユーザ作成

### 管理者でサインイン
管理者で入力したパスワードを使ってサインインします。



* Username or email: root
* Password: 初期設定で入力したパスワード

### ユーザの新規作成
トップ画面のメニュー[New user]を押下します。




###  ユーザ情報の入力
作成するユーザの情報を入力します。
* Name
* Username
* Email



### ユーザの作成

ボタン[Create user]を押下します。



### パスワードの再設定
(任意)作成したユーザのパスワードを再設定します。
作成したユーザにメールが届く場合はパスワードを設定できるため、本手順は不要です。

#### ユーザ情報の編集
作成したjenkinsユーザの`Edit`を押下します



#### パスワードの更新
パスワードを設定します。



#### ユーザ情報の更新確定

ボタン`Save changes`を押下します




## プロジェクト(リポジトリ)を作成

### サインイン
作成したユーザでサインインします。



### プロジェクトの作成開始

メニュー[New Project]を押下します



### プロジェクトの種類選択
タブ[Blank]を押下します



### 名前の決定
[Project name]を入力します。ここで設定した値がリポジトリのパスになります。



### 公開モードの決定

全体に公開するか、プライベートにするかを決定します。



### プロジェクトの作成

ボタン[Create project]を押下します。
