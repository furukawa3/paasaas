# gitbucketの設定



## ユーザ作成

### サインインを開始
右上の[Sign in]を押下



### 管理者でサインイン
User: root
Password: root
[Sign in]を行います。



### 管理者メニューを開く
メニューから[System administration]を選択します。




### ユーザを新規に作成
メニューから[New User]を押下します。




### 作成するユーザ情報を入力
新規に作成するユーザの情報を入力します。
以下の項目を正しく入力します。
* Username
* Password
* Full Name
* Mail Address




### 作成
ユーザ情報の入力が完了したら[Create user]を押下します。



### ユーザ作成確認
ユーザが作成されたことを一覧画面で確認します。




### 管理者サインアウト
メニューから[Sign out]を押下します。







## リポジトリを作成

### 作成したユーザでサインイン
先で作成したユーザでサインインします。



### リポジトリを作成開始
メニュー[New repository]を選択



### 作成するリポジトリの設定

Repository nameを入力します。


### リポジトリを作成
[Create repository]を押下します。
