# Jenkinsの設定

## 初期化

### トークンの確認
トークンを確認する

```python
!ansible -b -a 'docker exec -it jenkins cat /var/jenkins_home/secrets/initialAdminPassword' {target_host}
```

### トークンの入力
URLにアクセスするとトークンが求められるので`/var/jenkins_home/secrets/initialAdminPassword`調べて入力します。
（上を実行で確認できます）



### トークンの入力の確認
`Continue`ボタンを押下します。



### Install Pluginsを選択
[Install Suggested Plugins]を押下します。


### 初期プラグインのインストール
自動的にプラグインがインストールされるので待ちます。



### 管理者の作成

以下の項目を入力します。
* ユーザー名
* パスワード
* パスワードの確認
* フルネーム
* メールアドレス




### 管理者の入力完了

入力が完了したら`Save and Finish`を押下します。


### Jenkinsを開始
すべての作業が完了したら`Start using Jenkins`を押下します。




## プラグインの導入

### Jenkinsの管理
メニュー[Jenkinsの管理]を押下します。




### プラグインのインストール

連携させるサービスに応じたプラグインをインストールします。詳細は各結合手順書を参照




### 利用可能なプラグインの選択

[利用可能]タブを選択して、インストールするプラグインをチェックします。



### プラグインのインストールと再起動

チェックが完了したら[ダウンロードして再起動後にインストール]を押下します。



## ジョブの作成

### 新規ジョブ作成の開始

[新規ジョブ作成]を押下します。



### ジョブ名とスタイルの選択
[Enter an item name]にジョブ名を入力し、フリースタイルを選択します。



### 入力完了
入力が完了したらOKを押下します。



### ジョブの設定

連携させるサービスに応じジョブの設定を行う。詳細は各結合手順書を参照
